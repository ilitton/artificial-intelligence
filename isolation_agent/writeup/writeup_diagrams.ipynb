```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
heuristic_1 = pd.DataFrame({'Opponent': ['Random', 'MM_Open', 'MM_Center', 'MM_Improved', 'AB_Open', 'AB_Center', \
                                         'AB_Improved']*4,\
                            'Win Rate': [90.0, 90.0, 60.0, 80.0, 50.0, 60.0, 50.0, 90.0, 60.0, 80.0, 60.0, 50.0, 70.0,\
                                         60.0, 80.0, 70.0, 50.0, 60.0, 40.0, 40.0, 20.0, 60.0, 50.0, 60.0, 70.0, 40.0,\
                                         20.0, 40.0],\
                            'Player': ['AB_Improved']*7 + ['AB_Custom']*7 + ['MM_Improved']*7 + ['MM_Custom']*7})
```

```python
plt.figure(figsize=(10,7))

plt.plot(range(-10, 10), [68.6]*20, color='xkcd:azure', linewidth=4)
plt.plot(range(-10, 10), [67.1]*20, color='xkcd:tomato', linewidth=4)
plt.plot(range(-10, 10), [51.4]*20, color='xkcd:lightblue', linewidth=4)
plt.plot(range(-10, 10), [48.6]*20, color='xkcd:orangered', linewidth=4)

# plt.annotate(68.6, xy=(68.6, 10),\
#                  xytext=(11.5, 68.6), weight='bold', color='blue',size=16, transform=plt.gcf().transFigure)
plt.text(6.55, 70, "68.6", fontsize=14, weight='bold', color='xkcd:azure')
plt.text(6.55, 65, "67.1", fontsize=14, weight='bold', color='xkcd:tomato')
plt.text(6.55, 52, "51.4", fontsize=14, weight='bold', color='xkcd:lightblue')
plt.text(6.55, 48, "48.6", fontsize=14, weight='bold', color='xkcd:orangered')

sns.barplot(x="Opponent", y="Win Rate", hue="Player", data=heuristic_1, alpha=0.45, palette=['xkcd:azure',\
            'xkcd:tomato', 'xkcd:lightblue', 'xkcd:orangered'])
plt.xlabel("Opponent", fontsize=15, fontweight='bold')
plt.ylabel("Win Rate (%)", fontsize=15, fontweight='bold')
plt.title("Limit Opponents Move Heuristic", fontsize=20, fontweight='bold')

night_patch = mpatches.Patch(color='xkcd:azure', label='AB_Improved')
morning_patch = mpatches.Patch(color='xkcd:tomato', label='AB_Custom')
afternoon_patch = mpatches.Patch(color='xkcd:lightblue', label='MM_Improved')
evening_patch = mpatches.Patch(color='xkcd:orangered', label='MM_Custom')

plt.legend(handles=[night_patch, morning_patch, afternoon_patch, evening_patch], loc=1)

plt.ylim([0,100])
```
